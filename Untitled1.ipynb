```python
def run_supg(df, budget, target, qtype, delta=0.05, verbose=False):
    source = datasource.DFDataSource(df)
    sampler = ImportanceSampler()
    query = ApproxQuery(
            qtype=qtype,
            min_recall=target, min_precision=target, delta=delta,
            budget=budget
    )
    if qtype == 'rt':
        selector = RecallSelector(query, source, sampler, sample_mode='sqrt', verbose=False)
    elif qtype == 'pt':
        selector = ImportancePrecisionTwoStageSelector(query, source, sampler)
    else:
        raise NotImplementedError
    inds, threshold = selector.select()
    true_results = np.sum(source.lookup(inds))
    precision = true_results / len(inds)
    # print([sum(df['label']), np.sum(source.lookup(source.get_ordered_idxs()))])
    # print(true_results)
    assert(sum(df['label']) == np.sum(source.lookup(source.get_ordered_idxs())))
    recall = true_results / np.sum(source.lookup(source.get_ordered_idxs()))
    f1 = 2 * (precision * recall) / (precision + recall)
    return {
        'frame_indexes': inds,
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'selectivity': len(df[df['label'] == 1]) / len(df),
        'y_true': sum(df[df['label'] == 1]['label'])
    }, threshold
```

```python
def plot_f1_budget(df, subject):
    budget_values = np.arange(100, 2000, 100)
    f1_scores_rt = []
    f1_scores_pt = []
    precision_pt = []
    recall_rt = []
    recall_pt = []
    precision_rt = []
    for budget in budget_values:
        try:
            res_pt = run_supg(df, budget, 0.95, 'pt')
            res_rt = run_supg(df, budget, 0.95, 'rt')
            f1_scores_rt.append(res_rt['f1_score'])
            f1_scores_pt.append(res_pt['f1_score'])
            precision_pt.append(res_pt['precision'])
            precision_rt.append(res_rt['precision'])
            recall_rt.append(res_rt['recall'])
            recall_pt.append(res_pt['recall'])
        except:
            print(f'Error on budget: {budget}')
            precision_rt.append(0)
            recall_rt.append(0)
            precision_pt.append(0)
            recall_pt.append(0)
            f1_scores_pt.append(0)
            f1_scores_rt.append(0)
            
    fig, ax = plt.subplots(3, 1, figsize=(7, 7))
    ax[0].plot(budget_values, precision_pt, label='precision_pt', color='blue', marker='o')
    ax[0].plot(budget_values, recall_pt, label='recall_pt', color='red', marker='o')
    ax[0].set_title(f'precision vs budget for {subject}')
    ax[0].set_ylabel('precision')
    # ax[0].legend()
    ax[1].plot(budget_values, recall_rt, label='recall_rt', color='red', marker='o')
    ax[1].plot(budget_values, precision_rt, label='precision_rt', color='blue', marker='o')
    ax[1].set_title(f'recall vs budget for {subject}')
    ax[1].set_xlabel('Target')
    ax[1].set_ylabel('recall')
    # ax[1].legend()
    ax[2].plot(budget_values, f1_scores_rt, label='rt', color='blue', marker='o')
    ax[2].plot(budget_values, f1_scores_pt, label='pt', color='red', marker='o')
    ax[2].set_title(f'F1 Score vs budget for {subject}')
    ax[2].set_xlabel('Target')
    ax[2].set_ylabel('F1 Score')
    ax[2].legend()
    fig.tight_layout()
    plt.show()
```

```python
plot_f1_budget(dfs['dog'], 'dog')
```

```python
def plot_proxy_frame(df, query_type, budget=10000, recall_target=0.9, delta=0.05):
    res, threshold = run_supg(df, budget, recall_target, query_type, delta=delta)
    print(f'Threshold = {threshold}')
    print(f'Count above threshold = {len(df[df['proxy_score'] >= threshold])}')
    fig, ax = plt.subplots(1, 1, figsize=(10, 7))
    colors = ['red' if label == 1 else 'blue' for label in df['label']]
    ax.bar(df['id'], df['proxy_score'], color=colors)
    ax.axhline(y=threshold, color='green', linestyle='--', linewidth=2, label=f"threshold returned for {query_type}")
    ax.set_xlabel('Frames (id)')
    ax.set_ylabel('Proxy Score')
    ax.set_title('Proxy Score by Frame ID')
    # ax.legend()
    plt.show()
```

```python
plot_proxy_frame(dfs['dog'], 'rt')
```

```python
def plot_f1_target(df, subject):
    target_values = np.arange(0.7, 0.99, 0.05)
    f1_scores_rt = []
    f1_scores_pt = []
    precision_pt = []
    precision_rt = []
    recall_rt = []
    recall_pt = []
    for target in target_values:
        try:
            res_pt, _ = run_supg(df, 10000, target, 'pt')
            res_rt, _ = run_supg(df, 10000, target, 'rt')
            f1_scores_rt.append(res_rt['f1_score'])
            f1_scores_pt.append(res_pt['f1_score'])
            precision_pt.append(res_pt['precision'])
            precision_rt.append(res_pt['recall'])
            recall_rt.append(res_rt['recall'])
            recall_pt.append(res_rt['precision'])
        except:
            print(f'Error on target: {target}')
            f1_scores_pt.append(0)
            f1_scores_rt.append(0)
            
    fig, ax = plt.subplots(3, 1, figsize=(7, 7))
    ax[0].plot(target_values, precision_pt, label='precision_pt', color='blue', marker='o')
    ax[0].plot(target_values, recall_pt, label='recall_pt', color='red', marker='o')
    ax[0].set_title(f'precision query scores vs Target for {subject}')
    ax[0].set_ylabel('precision')
    ax[0].legend()
    ax[1].plot(target_values, recall_rt, label='recall_rt', color='red', marker='o')
    ax[1].plot(target_values, precision_rt, label='precision_pt', color='blue', marker='o')
    ax[1].set_title(f'recall query scores vs Target for {subject}')
    ax[1].set_xlabel('Target')
    ax[1].set_ylabel('recall')
    ax[1].legend()
    ax[2].plot(target_values, f1_scores_rt, label='rt', color='blue', marker='o')
    ax[2].plot(target_values, f1_scores_pt, label='pt', color='red', marker='o')
    ax[2].set_title(f'F1 Score vs Target for {subject}')
    ax[2].set_xlabel('Target')
    ax[2].set_ylabel('F1 Score')
    ax[2].legend()
    fig.tight_layout()
    plt.show()
```

```python
def normalise_df(df):
    df_normalised = df.copy()

    scaler = MinMaxScaler()
    df_normalised['proxy_score'] = scaler.fit_transform(df[['proxy_score']])

    return df_normalised
```

```python
df_dogs_normal = normalise_df(dfs['dog'])
plot_f1_target(df_dogs_normal, 'dog')
```
